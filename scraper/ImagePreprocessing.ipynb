```python
def trimImages(allSides, top, left, right, bottom):
    imagelist = os.listdir('picturesC')
    for i in range(1,len(imagelist)):
        crop_img = cv2.imread("picturesC/" + str(i))
        dimensions = crop_img.shape[0:2]
        print(dimensions)
        if(i in allSides):
            crop_img = crop_img[20:dimensions[0]-20, 20:dimensions[1]-20]
            print('all')
        else:
            if(i in top):
                crop_img = crop_img[20:dimensions[0], 0:dimensions[1]]
                dimensions = crop_img.shape[0:2]
                print('here')
            if(i in bottom):
                crop_img = crop_img[0:dimensions[0]-20, 0:dimensions[1]]
                dimensions = crop_img.shape[0:2]
                print('there')
            if(i in left):
                crop_img = crop_img[0:dimensions[0], 20:dimensions[1]]
                dimensions = crop_img.shape[0:2]
                print('beep')
            if(i in right):
                crop_img = crop_img[0:dimensions[0], 0:dimensions[1]-20]
                dimensions = crop_img.shape[0:2]
                print('boop')
        print(crop_img.shape[0:2])
        cv2.imwrite('ModifiedImages/' + str(i) +'.png',crop_img) 
            
```

```python
def resizeImageFolder(sizeX, sizeY, outputSuffix, directory, outputDirectory):
    
    imagelist = os.listdir(directory)
    for i in range(1,len(imagelist)):
        crop_img = Image.open(directory + '/' + str(imagelist[i]))
        size = (sizeX, sizeY)
        fit_and_resized_image = ImageOps.fit(crop_img, size, Image.ANTIALIAS)
        rgb_im = fit_and_resized_image.convert('RGB')
        #gray = ImageOps.grayscale(rgb_im)
        rgb_im.save(outputDirectory + '/' + str(i) + outputSuffix)
                                
        
                           
```

```python
resizeImageFolder(160, 160,'.png','ChinesePorcelain', 'ChinesePorcelainColor')
```

```python
def createRotations(outputSuffix, directory, outputDirectory):
    imagelist = os.listdir(directory)
    for i in range(len(imagelist)):
        img = Image.open(directory + '/' + str(imagelist[i]))
        rotation = 90
        for j in range(3):
            rotate_img = img.rotate(rotation)
            rotate_img.save(outputDirectory + '/' + str(i) + '_' + str(rotation) + outputSuffix)
            rotation = rotation + 90
```

```python
createRotations('.png','ChinesePorcelainColor', 'ChinesePorcelainColor')
```
